```python
#Declaração das bibliotecas
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import quad
```

#### 1. Escreva uma função para plotar os gráficos e as derivadas analíticas de:

$y = x^3 + x^2 - 10 x + 10, ~x \in [-5,5]$

```python
# Define a função y = x^3 + x^2 - 10x + 10
def func(x):
    return x**3 + x**2 - 10*x + 10

# Define a derivada analítica d1y = 3x^2 + 2x - 10
def deriv(x):
    return 3*x**2 + 2*x - 10
    
# Define o número de pontos 
N = 50

# Pontos a serem plotados no gráfico
x = np.linspace(-5,5,50)
y = func(x)
d1y = deriv(x)

#Gráfico da função y = x^3 + x^2 - 10x + 10 com sua primeira derivada analítica d1y = 3x^2 + 2x - 10
plt.plot(x,y, "b", label= "Função y")
plt.plot(x,d1y, "r", label="1ª derivada de y")
plt.legend()
plt.title("Função y e derivada d1y")
plt.show()
```

#### 2. Calcule numericamente as derivadas das funções acima usando as derivadas progressiva e central. Faça gráficos comparando os resultados obtidos com as derivadas exatas.

```python
# cálculo da 1a. derivada progressiva (erro da ordem de dx^2)
d1yp = [(y[i+1]-y[i])/(x[i+1]-x[i]) for i in range(0,N-1)]

# cálculo da 1a. derivada de ponto central (erro da ordem de dx^3)
d1yc = [(y[i+1]-y[i-1])/(x[i+1]-x[i-1]) for i in range(1,N-1)]

# gráfico das derivadas progressiva e central
plt.plot(x,d1y,"bo-", label= "1ª derivada analítica")
plt.plot(x[0:(N-1)],d1yp,'k--',label="1ª derivada progressiva")
plt.plot(x[1:(N-1)],d1yc,'r', label= "1ª derviada central")
plt.legend()
plt.title("1ª derivada")
plt.show()
```

#### 3. Escreva um código para calcular o erro nas derivadas em cada ponto $\Delta_i = |f_{num}(x_i)- f_{exato}(x_i)|$ e o erro integrado $\Delta_{tot} = \sum_i \Delta_i$. Faça um gráfico do erro integrado em função de $\Delta x$ para cada uma das derivadas das funções do problema 1. Repita este cálculo para as derivadas progressiva e central. Discuta os resultados em função do truncamento na série de Taylor que deu origem a aproximação da derivada.

```python
#Cálculo do erro para as derivadas progressiva e central da função y = x^3 + x^2 - 10x + 10

# cálculo do erro em cada ponto x_i
err_d1yp = np.abs(d1yp-d1y[0:(N-1)]) 
err_d1yc = np.abs(d1yc-d1y[1:(N-1)])

# cálculo do erro integrado
int_err_d1yp = np.sum(err_d1yp)
int_err_d1yc = np.sum(err_d1yc)

# gráfico 1: 1a. derivada
plt.plot(x[0:(N-1)],err_d1yp,'k--',label="Erro na 1ª derivada progressiva")
plt.plot(x[1:(N-1)],err_d1yc,'r',label= "Erro na 1ª derviada central")  
plt.legend()
plt.title('Erro na 1ª derivada')
plt.show()

# exibição dos erros integrados
print("Espaçamento dx entre x[i] e x[i+1]: %5.3f" % (x[1]-x[0]))
print("Erro integrado na 1ª derivada progressiva: %5.3f" % int_err_d1yp)
print("Erro integrado na 1ª derivada de ponto central: %5.3f" % int_err_d1yc)
```
