```python
#Declaração das bibliotecas
import matplotlib.pyplot as plt
import numpy as np
```

#### 2 - Resolva o problema 1.4 do livro-texto.

```python
dt = 0.2          
# numero de passos no tempo
nt = 500   
# meia vida do elemento A
tau_a = 2  
# meia-vida do elemnto B
tau_b = 1  
# Quantidade inicial de NA
N0_a = 100  
# Quantidade inicial de NB
N0_b = 100  

# criação do vetor para o tempo (carregado com zeros)
time = np.zeros(nt)   

# criação do vetor para os valores NA-Euler 
decay_a = np.zeros(nt)  
decay_a[0] = N0_a

# criação do vetor para os valores NB-Euler
decay_b = np.zeros(nt)   
decay_b[0] = N0_b

# criação dos vetores para os valores NA e NB Analitico
theory_a = np.zeros(nt)
theory_a[0] = N0_a

theory_b = np.zeros(nt)
theory_b[0] = N0_b

# iteração preenchimento de tempo, valores NA e NB de Euler e analitico
for i in range(1,nt):
    time[i] = time[i-1] + dt 
    theory_a[i] = N0_a*np.exp(-time[i]/tau_a)
    decay_b[i] = theory_a[i-1] *(dt/tau_a) + decay_b[i-1]*(1-dt/tau_b)
    theory_b[i] = N0_a/((tau_a/tau_b)-1)*(np.exp(-time[i]/tau_a) - np.exp(-time[i]/tau_b))+N0_b*np.exp(-time[i]/tau_b)


plt.plot(time, theory_a,'b-o',time,decay_b,'-r',time,theory_b,'k--')
plt.title('Decaimento Radioativo para dois tipos de Núcleos', fontsize=16)
plt.xlabel('Tempo')
plt.ylabel('N(t)')
plt.text(8,60,'tau_a=2 \ntau_b=1')
plt.legend(['Teórico Na','Numérico Nb','Teórico Nb'])
plt.axis([0, 10, -3, 110])
plt.show()
```

#### 3 - Resolva o problema 1.5 do livro-texto.

```python
dt = 0.05  
# numero de passos no tempo
T = 1000  
# constante de tempo
tau = 2  
# numero inicial de nucleos do elemento A
N0_a = 0  
# numero inicial de nucleos do elemento B
N0_b = 100  

# vetor de tempo (preenchido com zeros)
time = np.zeros(T)  
tmax = ( T -1 )*dt

# vetor de quantidade de nucleo A
decay_a = np.zeros(T)  
decay_a[0] = N0_a       # atribuição de valor para primeira posição

# vetor de quantidade de nucleo B
decay_b = np.zeros(T)   
decay_b[0] = N0_b       # atribuição de valor para primeira posição      


# carregamento dos vetores com valores pelo metodo de Euler       
for i in range(1,T):
    time[i] = time[i-1]+dt
    decay_a[i]=decay_a[i-1]+dt*(decay_b[i-1]-decay_a[i-1])/tau
    decay_b[i]=decay_b[i-1]+dt*(decay_a[i-1]-decay_b[i-1])/tau
    


plt.plot(time,decay_a,'b--',time,decay_b,'r--')
plt.title('Ressonância entre dois tipos de Núcleos', fontsize=16)
plt.xlabel('Tempo')
plt.ylabel('N(t)')
plt.legend(['N_a(t)','N_b(t)'])
plt.axis([0, 20, -3, 110])
plt.show()
```
